# Data Analysis

```python
corr_matrix=data.corr(numeric_only=True)
print(corr_matrix)
```

```python
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
corr_matrix=data.corr(numeric_only=True)
```

```python


# Filter strong correlations
strong_corr = corr_matrix[(corr_matrix > 0.5) | (corr_matrix < -0.5)]

# Drop rows/columns with all NaNs
strong_corr = strong_corr.dropna(how='all').dropna(axis=1, how='all')
print(strong_corr)
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(strong_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Strong Correlations Only")
plt.show()
```
